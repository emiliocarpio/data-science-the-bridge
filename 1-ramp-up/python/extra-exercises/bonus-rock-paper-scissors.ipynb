```python
rock = 1
paper = 2
scissor = 3

def get_inputs():
    player1 = int(input("Choose 1[rock], 2[paper], 3[scissor]:"))
    player2 = int(input("Choose 1[rock], 2[paper], 3[scissor]:"))
    if (player1 != 1 and player1 != 2 and player1 != 3) or (player2 != 1 and player2 != 2 and player2 != 3):
        print("Only there are 3 possibilities: [1, 2, 3]")
        player1, player2 = get_inputs()
        return player1, player2    
    else:
        return player1, player2

def p1_is_rock(p2):
    if p2 == rock:
        print("Draw")
    elif p2 == paper:
        print("Player 2 wins")
    elif p2 == scissor:
        print("Player 1 wins")

def p1_is_paper(p2):
    if p2 == rock:
        print("Player 1 wins")
    elif p2 == paper:
        print("Draw")
    elif p2 == scissor:
        print("Player 2 wins")

def p1_is_scissor(p2):
    if p2 == rock:
        print("Player 2 wins")
    elif p2 == paper:
        print("Player 1 wins")
    elif p2 == scissor:
        print("Draw")

def check_winner(player1_choose, player2_choose):
    print("player1_choose:", player1_choose)
    print("player2_choose:", player2_choose)
    if player1_choose == rock:
        p1_is_rock(p2=player2_choose)
    elif player1_choose == paper:
        p1_is_paper(p2=player2_choose)
    elif player1_choose == scissor:
        p1_is_scissor(p2=player2_choose)

g, h = get_inputs()
check_winner(player1_choose=g, player2_choose=h)
```

```python
def rock_paper_scissors (rounds):
    """
                        ---What it does---
    This function is a rock-paper-scissors game. This function asks for the users input for his/her nickname of choice and the choice for the game (either rock, paper... etc.). Then tracks the number of victories and draws of each players and returns them.
                        ---What it needs---
        - The number of rounds to be played. It should be greater than 0 and an int.
    """

    def get_nick_name ():
        """
                            ---What it does---
        This function asks the users for his/her nickname of choice.
                            ---what it needs---
            - The user's input
                            ---What it returns---
        This function returns the nickname as a string.
        """

        print('How thee be named?')
        nick_name = input('> ')
        print(f'{nick_name}? Got it!')
        print()

        return nick_name

    def choice_generator(player, possibilities):
        """
                            ---What it does---
        This function allows the user to make the choice for the match (either rock, paper of scissors) by a dictionary of predefined options.
                            ---what it needs---
            - The player nickname (player)
            - A dictionary of valid moves (possibilities)
                            ---What it returns---
        This function returns the player's choice as a number (int).
        """

        print(f'\nReady {player}?')
        
        message = f'Make your choice {player}, remember to enter a number from 1 to 3!'
        choice = int(input(message))

        if choice > 3:
            print('Out of range! Choose again!')
            choice = int(input(message))
        
        print('You chose wisely\n')

        return choice

    def turn_sequence(player_1, player_2, choice_player_1, choice_player_2):
        """
                            ---What it does---
        This function creates the turn sequence for both players and determines the winner of the round.
                            ---what it needs---
            - The nickname for player_1 (player_1)
            - The nickname for player_2 (player_2)
            - Player_1's choice (choice_player_1)
            - Player_2's choice (choice_player_2) 
                            ---What it returns---
        This function returns the victor of the round (victory) as a string.
        """

        print("\nLet's get ready for rumble!\n")
        player_choice = possibilities[choice_player_1]
        player_choice_2 = possibilities[choice_player_2]

        if choice_player_1 < choice_player_2 and choice_player_2 != 3:
            print(f'\t{player_1} chose {player_choice}...\n\t{player_2} chose {player_choice_2}')
            print(f'\t{player_1} wins the round!')

            victory = 'player_1'
            
            return victory

        elif choice_player_2 < choice_player_1 and choice_player_2 != 3:
            print(f'\t{player_1} chose {player_choice}...\n\t{player_2} chose {player_choice_2}')
            print(f'\t{player_2} wins the round!')

            victory = 'player_2'
            
            return victory

        else:
            print(f'\t{player_1} chose {player_choice}...\n\t{player_2} chose {player_choice_2}')
            print(f'\t{player_1} lost? but {player_2} did not win? Everyone loses!')

            victory = 'draw'
            
            return victory

    def victory_count(player_1, player_2, victory):
        """
                            ---What it does---
        This functionj calculates who of the two is the ultimate winner of the match by adding the victories across the rounds.
                            ---what it needs---
            - The nickname for player_1 (player_1)
            - The nickname for player_2 (player_2)
            - The winner of the round (victory)
                            ---What it returns---
        This function returns the results of the round in a dictionary (match).
        """
        match = {player_1: [], player_2: []}
        player_2_wins = {}

        if victory == 'player_1':
            match[player_1].append(1)
            match[player_2].append(0)

        elif victory == 'player_2':
            match[player_1].append(0)
            match[player_2].append(1)

        else:
            match[player_1].append(0)
            match[player_2].append(0)
        
        return match
       

    def winner (player_1, player_2, match):
        """
                            ---What it does---
        This function
                            ---what it needs---
            - 
                            ---What it returns---
        This function returns
        """
        player_1_wins = sum(match[player_1])
        player_2_wins = sum(match[player_2])

        print(f'\n{player_1} won {player_1_wins} and {player_2} won {player_2_wins}')

        if player_1_wins > player_2_wins:
            print(f'\n\t{player_1.upper()} WINS!')

        elif player_1_wins < player_2_wins:
            print(f'\n\t{player_2.upper()} WINS!')
        
        else:
            print("It's a draw...\nHa ha!")

    possibilities = {1: 'paper', 2: 'rock', 3: 'scissors'}

    player_1 = get_nick_name()
    player_2 = get_nick_name()

    for e in range(rounds):

        choice_player_1 = choice_generator(possibilities = possibilities, player = player_1)
        choice_player_2 = choice_generator(possibilities = possibilities, player = player_2)

        victory = turn_sequence(player_1= player_1, player_2= player_2, choice_player_1 = choice_player_1, choice_player_2= choice_player_2)
        match = victory_count(player_1= player_1, player_2= player_2, victory= victory)
    
    winner (player_1= player_1, player_2= player_2, match= match)
```
